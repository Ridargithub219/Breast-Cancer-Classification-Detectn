Importing the Dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
print(breast_cancer_dataset['DESCR'])
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

1 --> Benign

0 --> Malignant

```python
data_frame.groupby('label').mean()    #by label alone
```

Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model using Training data

model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy*100)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy*100)
```

Scaling x

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
X_scaled=scale.fit_transform(X)
X_scaled
```

Before pca model testing ,train_test_split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)*100
```

```python
#import pca
from sklearn.decomposition import PCA
pca=PCA(0.95)
x_pca=pca.fit_transform(X_scaled)
x_pca
```

```python
#new train test split
from sklearn.model_selection import train_test_split
X_train_pca,X_test_pca,y_train,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=30)
modelpca=LogisticRegression(max_iter=1000)
modelpca.fit(X_train_pca,y_train)
modelpca.score(X_test_pca,y_test)*100
```

```python
#2 components
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
x_pca=pca.fit_transform(X_scaled)
x_pca
```

```python
X_scaled.shape
```

```python
x_pca.shape #perks :instead of computind data on 30 features we doing it on only 2 features
```

plotting pca

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(7,5))
plt.scatter(x_pca[:,0],x_pca[:,1],c=breast_cancer_dataset.target,cmap='plasma')
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
plt.show
```

```python
pca.components_
```

```python
#turnfor heatmap visualization
```

```python
import seaborn as sns
df_comp=pd.DataFrame(pca.components_,columns=breast_cancer_dataset.feature_names)
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```

Open and Read : This heatmap and the colorbar basically represnt the relationship between the varous features nd principle components itself.
There comes a need for this to do as with the great power of dimensionality reduction in pca comes the cost of NOT being able to easily Understand of what these components represent

```python
#new train testing componets reduced to 2
from sklearn.model_selection import train_test_split
X_train_pca,X_test_pca,y_train,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=30)
modelpca=LogisticRegression(max_iter=1000)
modelpca.fit(X_train_pca,y_train)
modelpca.score(X_test_pca,y_test)*100
```
